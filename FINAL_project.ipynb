## Problem 1. Position vs. time

## Problem 4. Will it bite you?

When the 'raptor is 1 meter away, it will try to bite you. If it doesn't bite you, it will try a second and (if necessary) a third time. If it doesn't get you on the third try, it will grow frustrated and stop chasing you. Hurray!

Suppose the first time it tries, there is a 20% chance it will bite you. If it misses and it needs to try a second time, there is only a 15% chance, and if it misses *that* time, there is only a 7% chance on the third try. 

If it bites you once, you're 'raptor food. If it misses all three times, you get away!

What is the probability that you will get away?


```python
# Your code here
import numpy as np
npeople=100000 #numbers will be generated howener many times specified here
nlived=0 #initial value
for i in range(0,npeople):
    first_chance = np.random.random() #generates percent for first try
    if first_chance>0.2:#b/c there is a 20% chance of being caught
        second_chance = np.random.random() #moves on to 2nd chance if condition for first chance is met
        if second_chance>0.15: #15% chance of being caught
            third_chance= np.random.random() #moves to 3rd if condition for first AND second are met 
            if third_chance>0.07: #7% chance of being caught
                #dont need the print statement here
                nlived+=1 #increases nlived variable by 1
print "There is a %s percent chance that you will get away." %((float(nlived)/npeople)*100)#gives the probability in a
#percent of how often you will get away
```
